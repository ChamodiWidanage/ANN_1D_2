```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Set the font to New Times Roman for Seaborn
sns.set(font='serif', style='whitegrid')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

### Peak pressure

```python
# Importing the dataset
dataset_1 = pd.read_excel('Peak_Pressure/PDataset_adj.xlsx')
dataset_1['Explosive type'] = dataset_1['Explosive type'].astype('category')
dataset_1.describe(include='category')
```

```python
dataset_1.corr()
```

```python
# Calculate the value
#corrMatrix1 = dataset_1.corr(method = 'kendall')
corrMatrix1 = dataset_1.corr(method = 'spearman')

# Set the font to New Times Roman for Seaborn
sns.set(font='serif', font_scale=2)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

# Create an instance of the PairGrid class.
grid = sns.PairGrid(data= dataset_1,
                    hue = "Explosive type",
                    vars = ["Explosive mass", "Standoff distance", "Peak incident pressure"],
                    palette=[ "red", "blue"],
                    diag_sharey=False,
                    corner=True,
                    height = 4)

#hue_marker={'Composition B':'o','TNT' :'^'}

# Map a scatter plot to the upper triangle
grid.map_offdiag(sns.scatterplot, alpha = 0.7,  style = dataset_1["Explosive type"], style_order = ["TNT", "Composition B"])

# Map a histogram to the diagonal
#grid.map_diag(sns.kdeplot)


# Map a density plot to the lower triangle
#grid = grid.map_lower(sns.kdeplot, cmap = 'Reds')
grid.add_legend(loc = (0.65,0.75))

# Add correlation coefficients to the plot
#for i, j in zip(*plt.np.tril_indices_from(grid.axes, k=-1)):
 #   r1 = corrMatrix1.iloc[i, j]
  #  grid.axes[i, j].annotate(f"K = {r1:.2f}", xy=(0.1, 1.02), xycoords='axes fraction', size = 18, color = 'Green')

# Add correlation coefficients to the plot
for l, m in zip(*plt.np.tril_indices_from(grid.axes, k=-1)):
    r2 = corrMatrix1.iloc[l, m]
    grid.axes[l, m].annotate(f"Spearman = {r2:.2f}", xy=(.2, 1.02), xycoords='axes fraction', size = 18, color = 'Green')


sns.despine()
plt.show()
```

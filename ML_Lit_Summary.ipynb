```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()

size = 0.3
vals = np.array([[1, 5, 0], [10, 4, 1], [1, 2, 0]])

cmap = plt.colormaps["tab20"]
inner_colors = cmap([0, 4, 6])
outer_colors = cmap([3, 9, 11, 3, 9, 11, 3, 9, 11])

ax.pie(vals.sum(axis=1),
       radius=1-size,
       colors=inner_colors,
       autopct= '%1.1f%%',
       pctdistance=0.77,
       wedgeprops=dict(width=size,edgecolor='w'))



ax.pie(vals.flatten(),
       radius=1,
       colors=outer_colors,
       autopct= '%1.1f%%',
       pctdistance=0.85,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.legend(labels = ["Regression", "Deep learning", "Regression +  Deep learning", "Blast load", "Blast load induced response", "Blast load + Blast load induced response"],
          loc="center left",
          facecolor = "white",
          bbox_to_anchor=(1, 0, 0.5, 1))

#ax.legend(labels = ["Load", "Load induced response", "Load + Load induced response", "Load", "Load induced response", "Load + Load induced response", "Load", "Load induced response", "Load + Load induced response"],
 #         loc="lower left",
  #        bbox_to_anchor=(1, 0, 0.5, 1))

ax.set(aspect="equal")

#plt.show()
plt.savefig('plot1.png', dpi = 600, bbox_inches = 'tight')
```

```python
years = ('2006', '2007', '2011', '2019', '2020', '2021', '2022', '2023', '2024')
m_types = {
    'Regression': np.array([0, 0, 1, 0, 1, 1, 3, 0, 0]),
    'Deep learning': np.array([1, 1, 0, 1, 1, 1, 4, 5, 1]),
    'Regression + Deep learning': np.array([0, 0, 0, 0, 0, 0, 2, 1, 0]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(9)

for m_type, count in m_types.items():
    p = ax.bar(years, count, width, label=m_type, bottom=bottom)
    bottom += count

    #ax.bar_label(p, label_type='center')

#ax.set_title('Number of penguins by sex')
ax.legend()
ax.set_facecolor("white")
ax.grid(visible=None, which='major', axis='both', color = "black")
plt.box(color = "black")

#plt.show()
plt.savefig('plot2.png', dpi = 600, bbox_inches = 'tight')
```

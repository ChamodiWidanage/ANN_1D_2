```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set font to Times New Roman and fontsize to 20
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 18

fig, ax = plt.subplots()

size = 0.3
vals = np.array([[2, 5, 0], [7, 5, 1], [2, 1, 0]])

cmap = plt.colormaps["tab20"]
inner_colors = cmap([0, 4, 6])
outer_colors = cmap([3, 9, 11, 3, 9, 11, 3, 9, 11])

# Define a custom autopct function to conditionally show percentages
def my_autopct(pct):
    return f'{pct:.0f}%' if pct != 0 else ''

ax.pie(vals.sum(axis=1),
       radius=1-size,
       colors=inner_colors,
       autopct= my_autopct,
       pctdistance=0.77,
       wedgeprops=dict(width=size,edgecolor='w'))



ax.pie(vals.flatten(),
       radius=1,
       colors=outer_colors,
       autopct= my_autopct,
       pctdistance=0.85,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.legend(labels = ["Regression", "Deep learning", "Regression +  Deep learning", "Blast load", "Blast load induced response", "Blast load + Blast load\ninduced response"],
          loc=[-0.5, -0.3],
          facecolor = "white",
          ncols = 2,
          columnspacing=0.5,
          frameon =False)

#ax.legend(labels = ["Load", "Load induced response", "Load + Load induced response", "Load", "Load induced response", "Load + Load induced response", "Load", "Load induced response", "Load + Load induced response"],
 #         loc="lower left",
  #        bbox_to_anchor=(1, 0, 0.5, 1))

ax.set(aspect="equal")

plt.savefig('plot1_revised.png', dpi = 600, bbox_inches = 'tight')
```

```python
years = ('2006', '2007', '2011', '2020', '2021', '2022', '2023', '2024')
m_types = {
    'Regression': np.array([0, 0, 1, 1, 1, 3, 0, 1]),
    'Deep learning': np.array([1, 1, 0, 1, 1, 3, 6, 0]),
    'Regression + Deep learning': np.array([0, 0, 0, 0, 0, 1, 2, 0]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(8)

for m_type, count in m_types.items():
    p = ax.bar(years, count, width, label=m_type, bottom=bottom)
    bottom += count

    #ax.bar_label(p, label_type='center')

ax.set_xticklabels(years, rotation=45)
ax.legend(fontsize = 15, loc = [0.01,0.6])
ax.set_facecolor("white")
#ax.grid(visible=None, which='major', axis='both', color = "black")
#plt.box(color = "black")


plt.savefig('plot2_revised.png', dpi = 600, bbox_inches = 'tight')
```

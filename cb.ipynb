```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])

plt.imshow(I)
colorbar = plt.colorbar(label='Peak Incident Pressure (kPa)')
# Adjust fontsize of ticks and label
colorbar.ax.tick_params(labelsize=20)  # Set fontsize for ticks
colorbar.set_label('Peak Incident Pressure (kPa)', fontsize=14)
```

### Importing the libraries

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```python
# Importing the dataset
dataset_tnt = pd.read_excel('Peak_Pressure/PDataset_adj_TNT.xlsx')
dataset_compb = pd.read_excel('Peak_Pressure/PDataset_adj_CompB.xlsx')
dataset_all_p = pd.read_excel('Peak_Pressure/PDataset_adj.xlsx')
dataset_all_i = pd.read_excel('Peak_Impulse/IDataset_adj.xlsx')
dataset_all_a = pd.read_excel('Arrival_time/ADataset_adj.xlsx')
```

```python
fig1, ax1 = plt.subplots(figsize = (9, 6))

ax1.set_facecolor('white')
n5, bins5, patches5 = plt.hist(x= [dataset_tnt['Explosive mass'],dataset_compb['Explosive mass']], bins= np.arange(dataset_tnt['Explosive mass'].min(), dataset_compb['Explosive mass'].max()+1),
                               color=['Red', 'Blue'],alpha=0.65, rwidth=0.85, label = ['TNT', 'Composition B'], stacked= 'True', align= 'mid')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['font.size'] = 18
plt.grid(visible= 'yes',axis='both', color = 'lightgrey')
plt.xlabel('Explosive Mass (kg)')
plt.ylabel('Frequency')
#plt.title('My Very Own Histogram')
#plt.text(23, 45, r'$\mu=15, b=3$')

plt.legend(loc=[0.4,1.03])
# Set a clean upper y-axis limit.
plt.ylim(0, 50)
plt.xlim(0, 26)

ax1.spines['left'].set_color('black')        
ax1.spines['bottom'].set_color('black')
ax1.spines['top'].set_color('black')        
ax1.spines['right'].set_color('black')

plt.show()
```

## Import libraries

```python
from pathlib import Path
import pandas as pd
import openpyxl
import math
```

## Access input and output directories 

```python
#Access input folder
input_dir1 = Path ("11series_tracers_pressure_excel")
print ("1",input_dir1)

#Access folders inside input folder
input_dir2 =  [folder for folder in input_dir1.iterdir() if folder.is_dir()]
print ("2",input_dir2)
```

## Filtering arrival time

```python
# Access each folder inside input folder
df = pd.DataFrame(columns = ['Explosive type','Explosive mass', 'Standoff distance', 'Time of arrival'])
for folder in input_dir2:
    print ("folder", folder)
    charge_mass = int(folder.stem[3:])/2
    
    # Make a list of data file names
    files = list(folder.rglob("*.xlsx"))
    
    for file in files:
        wrkbk = openpyxl.load_workbook(file)
        sh = wrkbk.active
        a=0
        stand_dis = int(file.stem[1:])/2
        
        # iterate through excel and display data
        for i in range(2, sh.max_row+1):
            cell_pressure = sh.cell (row = i, column = 2)
            if cell_pressure.value > 100000:
                print (file.stem)
                print ("pressure", cell_pressure.value)
                cell_time = sh.cell (row = i, column = 1)
                print ("time of arrival", cell_time.value)
                # append rows to an empty DataFrame
                new_row = {'Explosive type' : ['Composition B'],
                           'Explosive mass' : [charge_mass],
                           'Standoff distance' : [stand_dis],
                           'Time of arrival' : [cell_time.value]}
                new_row_data = pd.DataFrame(new_row)
                df = pd.concat([df, new_row_data], ignore_index = True)
                a=a+1
            if a == 1:
                break
```

```python
print (df)
```

## Create Arrival_time excel file

```python
with pd.ExcelWriter('ADataset1_TNT.xlsx', engine = "openpyxl", mode='a', if_sheet_exists='replace') as writer:  
    df.to_excel(writer, sheet_name = "Sheet1", index = False, header = True)
```
